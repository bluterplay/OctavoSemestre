```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## Simular números normales estándar

```python
delta = 0.5
T = 10000

X =[]
X.append(0)

for i in range(1,T):
    eps = np.random.uniform(-delta,delta)
    y = X[i-1]+eps
    alpha = min([(stats.norm.pdf(y))/(stats.norm.pdf(X[i-1])),1])
    u = np.random.uniform()
    if u <=alpha:
        X.append(y)
    else:
        X.append(X[i-1])

sns.distplot(X)
sns.distplot(np.random.normal(size=T))
plt.figure()
plt.plot(X)
```

## Simular Números $X\sim\exp(\lambda=2)$

```python
delta = 0.5
T = 10000

X =[]
X.append(0)

for i in range(1,T):
    eps = np.random.uniform(-delta,delta)
    y = X[i-1]+eps
    alpha = min([(stats.expon(scale=1/2).pdf(y))/(stats.expon(scale=1/2).pdf(X[i-1])),1])
    u = np.random.uniform()
    if u <=alpha:
        X.append(y)
    else:
        X.append(X[i-1])

sns.distplot(X)
sns.distplot(np.random.exponential(scale=1/2,size=T))
plt.figure()
plt.plot(X)
```


## Simular Números $F_{\nu_1=4,\nu_2=23}$

```python
delta = 0.5
T = 10000

X =[]
X.append(0)

for i in range(1,T):
    eps = np.random.uniform(-delta,delta)
    y = X[i-1]+eps
    alpha = min([(stats.f(4,23).pdf(y))/(stats.f(4,23).pdf(X[i-1])),1])
    u = np.random.uniform()
    if u <=alpha:
        X.append(y)
    else:
        X.append(X[i-1])

sns.distplot(X)
sns.distplot(np.random.f(4,23,size=T))
plt.figure()
plt.plot(X)
```
