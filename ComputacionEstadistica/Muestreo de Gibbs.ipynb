```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## Normal bivariada te√≥rica $\mathbf{\mu}=\begin{bmatrix} 0\\0\end{bmatrix}, \Sigma = \begin{bmatrix} 1& \frac{1}{2}\\\frac{1}{2}&1\\\end{bmatrix}$

```python
T = 1000
normal_multivariada = np.random.multivariate_normal([0,0],[[1,0.5],[0.5,1]],T)
```

```python
plt.scatter(normal_multivariada[:,0],normal_multivariada[:,1])
```

## Simulaci√≥n usando distribuciones condicionales
$p(x\vert y) = \mathcal{N}(\rho y,1-\rho^2)\\
p(y\vert x) = \mathcal{N}(\rho x,1-\rho^2)\\ 
$

```python
X = [1]
Y = [-1]
```

```python
for _ in range(T):
    X.append(np.random.normal(Y[-1]/2, np.sqrt(3/4)))
    Y.append(np.random.normal(X[-1]/2, np.sqrt(3/4)))
```

```python
plt.scatter(X, Y)
```

```python
sns.distplot(normal_multivariada[:,0])
sns.distplot(X)
```

```python
sns.distplot(normal_multivariada[:,1])
sns.distplot(Y)
```

```python
fig = plt.figure(figsize=(5,5),dpi=200)
ax = fig.add_subplot(111)
ax.scatter(X, Y,c='blue');
ax.plot(X, Y,c='red', marker='+',alpha=.12)


plt.scatter(normal_multivariada[:,0],normal_multivariada[:,1])
```
