## Modules

```python
import pandas as pd
from dirty_cat import SimilarityEncoder
```

## Data

```python
data= pd.read_csv("df_text.csv")
```

```python
data.head()
```

## Similarity encoder

```python
for n in [50,100,200]:
    df= data.copy()
    ls_columns= ["Var_"+str(x+1) for x in  range(n)]
    
    similarity_encoder = SimilarityEncoder(ngram_range=(2, 5),categories='most_frequent', n_prototypes= n)
    transformed_values = similarity_encoder.fit_transform(data[["Description"]])
    
    df[ls_columns]= transformed_values
    filename= "Data/dataSimilarityEnc"+str(n)+".csv"
    df.to_csv(filename, index_label=False)
```
